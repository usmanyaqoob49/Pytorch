```python
import pandas as pd
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/'
                 'machine-learning-databases'
                 '/breast-cancer-wisconsin/wdbc.data', header=None)
```

```python
df.head()
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder

X= df.loc[:,2:].values
y= df.loc[:,1].values

le= LabelEncoder()
y= le.fit_transform(y)

print('Classes Representation: ', le.classes_)
```

B is represented as 0 and M is represented as 1.

```python
le.transform(['B', 'M'])
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(
    X,
    y,
    test_size= 0.20,
    stratify= y,
    random_state= 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

### Pipeline:

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
```

```python
lr_pipeline= make_pipeline(
    StandardScaler(),
    PCA(n_components= 2),
    LogisticRegression())

lr_pipeline.fit(x_train, y_train)
predictions= lr_pipeline.predict(x_test)
accuracy= lr_pipeline.score(x_test, y_test)
print('Accuracy of Logistic Regression: ', accuracy)
```

### Stratified K-Fold Cross Validation:

Following is the Stratified K fold functinality written without using sklearn k fold cross validation module that offers Stratified K fold cross validation:

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold

stratified_kfolds= StratifiedKFold(n_splits= 10).split(x_train, y_train)
kfolds_accuracy_scores= []

for k_fold, (train_folds, test_fold) in enumerate(stratified_kfolds):
    lr_pipeline.fit(x_train[train_folds], y_train[train_folds])
    fold_score= lr_pipeline.score(x_train[test_fold], y_train[test_fold])
    kfolds_accuracy_scores.append(fold_score)

    print(f'Fold: {k_fold+1:02d}, '
          f'Class distr.: {np.bincount(y_train[train_folds])}, '
          f'Acc.: {fold_score:.3f}')
    
mean_acc = np.mean(kfolds_accuracy_scores)
std_acc = np.std(kfolds_accuracy_scores)
print(f'\nCV accuracy: {mean_acc:.3f} +/- {std_acc:.3f}')
```

Here is how we can do it with the sklearn cross validation module:

```python
from sklearn.model_selection import cross_val_score

scores= cross_val_score(estimator= lr_pipeline,
                                    X= x_train,
                                   y= y_train,
                                   cv= 10,
                                   n_jobs= -1 #advantage --> By n_jobs we can allow this module to use multiple CPUs available (by 1 it is using 1 CPU, by -1 it will use all the available CPUs)
                                  )

print(f'CV accuracy scores: {scores}')
print(f'CV accuracy: {np.mean(scores):.3f} '
      f'+/- {np.std(scores):.3f}')
```

## Diagnosing our ML models using Learning Curve:

To check how much they are generalizing by using overfitting and underfitting metrics.

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve

lr_pipeline= make_pipeline(StandardScaler(),
                           LogisticRegression(penalty= 'l2',
                                              max_iter= 10000)
                          )

train_sizes, train_scores, test_scores= learning_curve(estimator= lr_pipeline,
                                                       X= x_train,
                                                       y= y_train,
                                                       train_sizes= np.linspace(0.1, 1.0, 10),
                                                       cv= 10,
                                                       n_jobs= 1
                                                      )

training_mean= np.mean(train_scores, axis=1)
train_std= np.std(train_scores, axis= 1)
testing_mean= np.mean(test_scores, axis= 1)
test_std= np.std(test_scores, axis= 1)

plt.plot(train_sizes,
         training_mean,
         marker= 'o',
         markersize=5,
         label= 'Training Accuracy'
        )
plt.fill_between(train_sizes,
                 training_mean + train_std,
                 training_mean - train_std,
                 alpha=0.15, color='blue')

plt.plot(train_sizes, testing_mean,
         color='green', linestyle='--',
         marker='s', markersize=5,
         label='Validation accuracy')
plt.fill_between(train_sizes,
                 testing_mean + test_std,
                 testing_mean - test_std,
                 alpha=0.15, color='green')

plt.grid()
plt.xlabel('Number of training examples')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.ylim([0.8, 1.03])
plt.tight_layout()
plt.show()
```

**It fits very well after 250, byt after 400 it is kind of getting wider. That means overfitting is starting.**

## Addressing Under and Overfitting using Validation Curve:

So basically in learning curves we were seeing what happen with the test accuracy for using the different percentages of the data. But in the validation curve we see what happen with the test accuracu with using different value for a parameter.

```python
from sklearn.model_selection import validation_curve

parameter_rang= [0.001, 0.01, 0.1, 1.0, 10.0, 100.0]
train_scores, test_scores= validation_curve(
                            estimator= lr_pipeline,
                            X= x_train,
                            y= y_train,
                            param_name= 'logisticregression__C',#inverse Regulaization Parameter
                            param_range= parameter_rang,
                            cv= 10
)

train_mean= np.mean(train_scores, axis= 1)
train_std= np.std(train_scores, axis= 1)
test_mean= np.mean(test_scores, axis= 1)
test_std= np.std(test_scores, axis= 1)

plt.plot(parameter_rang, train_mean, 
         color='blue', marker='o', 
         markersize=5, label='Training accuracy')

plt.fill_between(parameter_rang, train_mean + train_std,
                 train_mean - train_std, alpha=0.15,
                 color='blue')

plt.plot(parameter_rang, test_mean, 
         color='green', linestyle='--', 
         marker='s', markersize=5, 
         label='Validation accuracy')

plt.fill_between(parameter_rang, 
                 test_mean + test_std,
                 test_mean - test_std, 
                 alpha=0.15, color='green')

plt.grid()
plt.xscale('log')
plt.legend(loc='lower right')
plt.xlabel('Inverse Regularization Parameter C')
plt.ylabel('Accuracy')
plt.ylim([0.8, 1.0])
plt.tight_layout()
```

As the parameter we were changing was Inverse Regularization Parameter, that means for large values there will be less regularization. So in the start for the small values of C, it was underfitting. Then at 0.01 seems to be optimal after that for the large values it started overfitting.

# Finetuning Models:

## Tunning Hyperparameter via Grid Search:

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

svc_pipeline= make_pipeline(StandardScaler(),
                            SVC(random_state= 42)
                           )

svc_parameter_range= [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0]
parameters_grid= [
    {'svc__C': svc_parameter_range,
     'svc__kernel': ['linear']},

    {'svc__C': svc_parameter_range,
     'svc__gamma': svc_parameter_range,
     'svc__kernel': ['rbf']}
    ]

grid_search= GridSearchCV(estimator= svc_pipeline,
                          param_grid= parameters_grid,
                          scoring= 'accuracy',
                          refit= True,
                          cv= 10
                         )
grid_search = grid_search.fit(x_train, y_train)
print(grid_search.best_score_)
print(grid_search.best_params_)
```

```python
best_classifier= grid_search.best_estimator_

print(f'Test accuracy: {best_classifier.score(x_test, y_test):.3f}')
```

## Tunning Hyperparameter via Randomized Search:

```python
from sklearn.model_selection import RandomizedSearchCV

svc_pipeline= make_pipeline(
    StandardScaler(),
    SVC(random_state= 42)
)
svc_parameter_range= [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0]
parameter_grid= [{'svc__C': svc_parameter_range,
               'svc__kernel': ['linear']},
              {'svc__C': svc_parameter_range,
               'svc__gamma': svc_parameter_range,
               'svc__kernel': ['rbf']}]

randomized_search= RandomizedSearchCV(estimator= svc_pipeline,
                                      param_distributions= parameter_grid, #it can take distribution and discrete values both
                                      scoring= 'accuracy',
                                      refit= True,
                                      n_iter= 20, #For trying 20 combos randomly - Main difference with Grid Search
                                      cv= 10,
                                      random_state= 42,
                                      n_jobs= -1
                                     )

randomized_search= randomized_search.fit(x_train, y_train)
print('Best Score: ', randomized_search.best_score_)
print('Best Parameter values: ', randomized_search.best_params_)
```

## Tunning Hyperparameter via Halving Randomized Search:

```python
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingRandomSearchCV
```

```python
halving_randomized_search= HalvingRandomSearchCV(
    estimator= svc_pipeline,
    param_distributions= parameter_grid,
    n_candidates= 'exhaust',
    resource= 'n_samples',
    factor= 1.5,
    random_state= 42,
    n_jobs= -1
)

halvingrandomizedsearch= halving_randomized_search.fit(x_train, y_train)
print('Best Accuracy Score: ', halvingrandomizedsearch.best_score_)
print('Best Parameters: ', halvingrandomizedsearch.best_params_)
```

```python
clf = halvingrandomizedsearch.best_estimator_
print(f'Test accuracy: {halvingrandomizedsearch.score(x_test, y_test):.3f}')
```

# Algorithm Selection with Nested Cross Validation:

```python
svc_parameter_range= [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0]
svc_parameter_grid= [{'svc__C': svc_parameter_range,
               'svc__kernel': ['linear']},
              {'svc__C': svc_parameter_range,
               'svc__gamma': svc_parameter_range,
               'svc__kernel': ['rbf']}]
svcgridsearch= GridSearchCV(
    estimator= svc_pipeline,
    param_grid= svc_parameter_grid,
    scoring= 'accuracy',
    cv= 2 #Outer K-Fold: To get best values for the hyperparameter
)
svc_scores= cross_val_score(estimator= svcgridsearch,
                            X= x_train,
                            y= y_train,
                            scoring= 'accuracy',
                            cv= 5 #Inner K-Fold: To check how model performs so that we can select best model      
                           )
print(f'CV accuracy SVC: {np.mean(svc_scores):.3f} '
      f'+/- {np.std(svc_scores):.3f}')

                            
from sklearn.tree import DecisionTreeClassifier
dtgridsearch= GridSearchCV(
    estimator= DecisionTreeClassifier(random_state= 42),
    param_grid= [{'max_depth': [1, 2, 3, 4, 5, 6, 7, None]}],
    scoring= 'accuracy',
    cv= 2 #Outer K-Fold
)
dt_scores= cross_val_score(estimator= dtgridsearch,
                           X= x_train,
                           y= y_train,
                           scoring= 'accuracy',
                           cv= 5 #Inner K-fold
                          )
print(f'CV accuracy DT: {np.mean(dt_scores):.3f} '
      f'+/- {np.std(dt_scores):.3f}')
```

**See, we can clearly select the best model that performed/generalized well with K-Fold Cross Validation with Best Hyperparameters**
